```python
import csv
import random
import mplcursors
def read_csv(filename):
    points = {}
    names = {}

    with open(filename, newline='') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            combo = row['Combination']
            costA = float(row['Cost_AgentA'])
            costB = float(row['Cost_AgentB'])
    
            if combo not in points:
                points[combo] = []
            if combo not in names:
                names[combo] = []
                names[combo].append(row['Selection'])
                names[combo].append(row['Crossover'])
                names[combo].append(row['Mutation'])
                
            points[combo].append([costA, costB])
            
    
    #print(points)
    print(names)

    return points, names

def generate_color(points):
    colors = {}
    for key in points:
        colors[key] = []
        colors[key].append(random.random())
        colors[key].append(random.random())
        colors[key].append(random.random())
    #print(colors)
    return colors



def draw_combined(points, colors, name):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(36, 12))
    all_scatters = []

    # ---- 左邊：畫散點圖 ----
    for key in points:
        x = [p[0] for p in points[key]]
        y = [p[1] for p in points[key]]
        ax1.scatter(x, y, color=colors[key], s=50)
        

    
    ax1.set_title("Line Chart Example")
    ax1.set_xlabel("Cost Agent A")
    ax1.set_ylabel("Cost Agent B")
    ax1.legend()

    # ---- 右邊：畫表格 ----
    ax2.axis('off')

    # 畫表頭
    ax2.text(0, 0, "Color", va='center', ha='left', fontsize=12, fontweight='bold')
    ax2.text(1, 0, "Selection", va='center', ha='left', fontsize=12, fontweight='bold')
    ax2.text(2, 0, "Crossover", va='center', ha='left', fontsize=12, fontweight='bold')
    ax2.text(3, 0, "Mutation", va='center', ha='left', fontsize=12, fontweight='bold')

    count = 1
    for key in points:
        # 畫點（左側）
        ax2.scatter(0, -count, color=colors[key], s=100)
        # 顯示文字（右側）
        ax2.text(1, -count, name[key][0], va='center', ha='left',fontsize=12)
        ax2.text(2, -count, name[key][1], va='center', ha='left',fontsize=12)
        ax2.text(3, -count, name[key][2], va='center', ha='left',fontsize=12)
        count += 1

    # 畫橫線（含表頭）
    for i in range(count + 1):
        ax2.hlines(-i + 0.5, -0.5, 3.5, colors='black', linewidth=0.5)

    # 畫直線（兩欄）
    ax2.vlines(-0.5, 0.5, -count + 0.5, colors='black', linewidth=0.5)
    ax2.vlines(0.5, 0.5, -count + 0.5, colors='black', linewidth=0.5)
    ax2.vlines(1.5, 0.5, -count + 0.5, colors='black', linewidth=0.5)
    ax2.vlines(2.5, 0.5, -count + 0.5, colors='black', linewidth=0.5)
    ax2.vlines(3.5, 0.5, -count + 0.5, colors='black', linewidth=0.5)
    plt.savefig("image.jpg", dpi=1000, bbox_inches='tight')
    plt.tight_layout()
    plt.show()


def draw_combined_v2(points, colors, name):
    # ---- 圖一：散點圖 ----
    fig1, ax1 = plt.subplots(figsize=(18, 12))

    for key in points:
        x = [p[0] for p in points[key]]
        y = [p[1] for p in points[key]]
        ax1.scatter(x, y, color=colors[key], s=50, label=key)

    ax1.set_title("Line Chart Example")
    ax1.set_xlabel("Cost Agent A")
    ax1.set_ylabel("Cost Agent B")
    ax1.legend()
    plt.tight_layout()
    fig1.savefig("scatter_plot.jpg", dpi=1000, bbox_inches='tight')
    plt.close(fig1)  # 關閉圖表，避免干擾下一張圖

    # ---- 圖二：表格圖 ----
    fig2, ax2 = plt.subplots(figsize=(18, 12))
    ax2.axis('off')

    # 畫表頭
    ax2.text(0, 0, "Color", va='center', ha='left', fontsize=12, fontweight='bold')
    ax2.text(1, 0, "Selection", va='center', ha='left', fontsize=12, fontweight='bold')
    ax2.text(2, 0, "Crossover", va='center', ha='left', fontsize=12, fontweight='bold')
    ax2.text(3, 0, "Mutation", va='center', ha='left', fontsize=12, fontweight='bold')

    count = 1
    for key in points:
        ax2.scatter(0, -count, color=colors[key], s=100)
        ax2.text(1, -count, name[key][0], va='center', ha='left', fontsize=12)
        ax2.text(2, -count, name[key][1], va='center', ha='left', fontsize=12)
        ax2.text(3, -count, name[key][2], va='center', ha='left', fontsize=12)
        count += 1

    # 畫橫線（含表頭）
    for i in range(count + 1):
        ax2.hlines(-i + 0.5, -0.5, 3.5, colors='black', linewidth=0.5)

    # 畫直線
    ax2.vlines(-0.5, 0.5, -count + 0.5, colors='black', linewidth=0.5)
    ax2.vlines(0.5, 0.5, -count + 0.5, colors='black', linewidth=0.5)
    ax2.vlines(1.5, 0.5, -count + 0.5, colors='black', linewidth=0.5)
    ax2.vlines(2.5, 0.5, -count + 0.5, colors='black', linewidth=0.5)
    ax2.vlines(3.5, 0.5, -count + 0.5, colors='black', linewidth=0.5)

    plt.tight_layout()
    fig2.savefig("table_plot.jpg", dpi=1000, bbox_inches='tight')
    plt.close(fig2)
```
